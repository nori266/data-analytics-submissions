```python
# correlations
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(20, 16))
heatmap = sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Heatmap', pad=20)
plt.tight_layout()
```

```python
# select categorical columns
categorical_cols = df.select_dtypes('object').columns
df[categorical_cols].describe()
```

```python
# distribution of severity
df.severity.value_counts(normalize=True)
```

```python
# cross-tabulations between severity and the accident type
pd.crosstab(df['severity'], df['accident_type'], normalize="columns")
```

```python
# the basic rate of traffic accidents that happen during hail
basic_ratio_of_hail = df.weather.value_counts(normalize=True)['raesade']
basic_ratio_of_hail
```

```python
ct = pd.crosstab(df['weather'], df['accident_type'], normalize="columns")
ct
```

```python
# What accident types are more associated with weather conditions (hail - raesade)?
# This association might mean increased risk of those types of accidents during hail.
significant_cols = ct.columns[ct.loc['raesade'] > basic_ratio_of_hail]
filtered_ct = ct[significant_cols]
filtered_ct
```

Observations:
- 15% of head-on collisions are fatal ones
- 15% of accidents are animal-involved, and 85% of those involve moose
- Looks like there's more risk to have an animal accident if there's hail
